# Transform ECR Data

```python
# Load ecr data for each year
start = 2012
end = 2024
for year in range(start, end):
    globals()['ecr' + str(year)] = pd.read_csv('FantasyPros_' + str(year) + '_Draft_ecr_ppr.csv')
    
# Transform ecr data
for year in range(start, end):
    # Rename columns
    globals()['ecr' + str(year)] = globals()['ecr' + str(year)].rename(columns={'AVG.': 'AVG', 
                                                                                'STD.DEV': 'STD_DEV', 
                                                                                'ECR VS. ADP': 'ECR_VS_ADP',
                                                                                'PLAYER NAME': 'PLAYER'})
    # Make new column 'POSRK' that is the numeric value of the 'POS' column
    globals()['ecr' + str(year)]['POSRK'] = globals()['ecr' + str(year)]['POS'].str.extract('(\d+)').astype(int)
    # Remove the digits from the 'POS' column
    globals()['ecr' + str(year)]['POS'] = globals()['ecr' + str(year)]['POS'].str.replace('\d+', '')
    # Remove columns
    globals()['ecr' + str(year)] = globals()['ecr' + str(year)].drop(columns=['TEAM'])
```

# Transform Fantasy Stats Data

```python
# Load fantasypros stats data for each year
start = 2012
end = 2024
for year in range(start, end):
        # Quarterback
    globals()['stats' + str(year) + '_qb'] = pd.read_csv('FantasyPros_Fantasy_Football_Statistics_QB_' + str(year) + '.csv')
    # Keep only the columns we need
    globals()['stats' + str(year) + '_qb'] = globals()['stats' + str(year) + '_qb'][['Player', 'FPTS', 'FPTS/G']]
    # Remove parantheses from player names
    globals()['stats' + str(year) + '_qb']['Player'] = globals()['stats' + str(year) + '_qb']['Player'].str.replace(r"\(.*\)","")
        # Running back
    globals()['stats' + str(year) + '_rb'] = pd.read_csv('FantasyPros_Fantasy_Football_Statistics_RB_' + str(year) + '.csv')
    # Keep only the columns we need
    globals()['stats' + str(year) + '_rb'] = globals()['stats' + str(year) + '_rb'][['Player', 'FPTS', 'FPTS/G']]
    # Remove parantheses from player names
    globals()['stats' + str(year) + '_rb']['Player'] = globals()['stats' + str(year) + '_rb']['Player'].str.replace(r"\(.*\)","")
        # Wide receiver
    globals()['stats' + str(year) + '_wr'] = pd.read_csv('FantasyPros_Fantasy_Football_Statistics_WR_' + str(year) + '.csv')
    # Keep only the columns we need
    globals()['stats' + str(year) + '_wr'] = globals()['stats' + str(year) + '_wr'][['Player', 'FPTS', 'FPTS/G']]
    # Remove parantheses from player names
    globals()['stats' + str(year) + '_wr']['Player'] = globals()['stats' + str(year) + '_wr']['Player'].str.replace(r"\(.*\)","")
        # Tight end
    globals()['stats' + str(year) + '_te'] = pd.read_csv('FantasyPros_Fantasy_Football_Statistics_TE_' + str(year) + '.csv')
    # Keep only the columns we need
    globals()['stats' + str(year) + '_te'] = globals()['stats' + str(year) + '_te'][['Player', 'FPTS', 'FPTS/G']]
    # Remove parantheses from player names
    globals()['stats' + str(year) + '_te']['Player'] = globals()['stats' + str(year) + '_te']['Player'].str.replace(r"\(.*\)","")
        # Kicker
    globals()['stats' + str(year) + '_k'] = pd.read_csv('FantasyPros_Fantasy_Football_Statistics_K_' + str(year) + '.csv')
    # Keep only the columns we need
    globals()['stats' + str(year) + '_k'] = globals()['stats' + str(year) + '_k'][['Player', 'FPTS', 'FPTS/G']]
    # Remove parantheses from player names
    globals()['stats' + str(year) + '_k']['Player'] = globals()['stats' + str(year) + '_k']['Player'].str.replace(r"\(.*\)","")
        # Defense
    globals()['stats' + str(year) + '_dst'] = pd.read_csv('FantasyPros_Fantasy_Football_Statistics_DST_' + str(year) + '.csv')
    # Keep only the columns we need
    globals()['stats' + str(year) + '_dst'] = globals()['stats' + str(year) + '_dst'][['Player', 'FPTS', 'FPTS/G']]
    # Remove parantheses from player names
    globals()['stats' + str(year) + '_dst']['Player'] = globals()['stats' + str(year) + '_dst']['Player'].str.replace(r"\(.*\)","")
```

```python
# Concatenate stats data based on year
start = 2012
end = 2024
for year in range(start, end):
    globals()['stats' + str(year)] = pd.concat([globals()['stats' + str(year) + '_qb'], 
                                                globals()['stats' + str(year) + '_rb'], 
                                                globals()['stats' + str(year) + '_wr'], 
                                                globals()['stats' + str(year) + '_te'], 
                                                globals()['stats' + str(year) + '_k'], 
                                                globals()['stats' + str(year) + '_dst']])
    # Add 'Year' column
    globals()['stats' + str(year)]['Year'] = year
    # Reset index
    globals()['stats' + str(year)] = globals()['stats' + str(year)].reset_index(drop=True)

# Convert 'Player' and 'PLAYER' columns to strip whitespace
for year in range(start, end):
    globals()['stats' + str(year)]['Player'] = globals()['stats' + str(year)]['Player'].str.strip()
    globals()['ecr' + str(year)]['PLAYER'] = globals()['ecr' + str(year)]['PLAYER'].str.strip()

# Now you can merge
for year in range(start, end):
    globals()['df' + str(year)] = pd.merge(globals()['ecr' + str(year)], globals()['stats' + str(year)], how='outer', left_on='PLAYER', right_on='Player')
    # Remove 'Player' column
    globals()['df' + str(year)] = globals()['df' + str(year)].drop(columns=['Player'])
```
